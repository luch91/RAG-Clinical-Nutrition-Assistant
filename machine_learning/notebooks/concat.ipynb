```python
import csv

input_file = "nutrient_query_qwen.csv"
output_file = "nutrient_query_qwen_clean.csv"

with open(input_file, "r", encoding="utf-8", newline="") as infile, \
     open(output_file, "w", encoding="utf-8", newline="") as outfile:

    reader = csv.reader(infile)
    writer = csv.writer(outfile, quoting=csv.QUOTE_MINIMAL)

    for row in reader:
        cleaned_row = []
        for field in row:
            if ',' in field and not (field.startswith('"') and field.endswith('"')):
                # Wrap field in quotes if it contains a comma and isn't already quoted
                cleaned_row.append(field)
            else:
                cleaned_row.append(field)
        writer.writerow(cleaned_row)

print(f"Cleaned CSV saved to {output_file}")
```

```python
import csv

input_file = "nutrient_query_qwen.csv"
output_file = "nutrient_query_qwen_clean.csv"

with open(input_file, "r", encoding="utf-8", newline="") as infile, \
     open(output_file, "w", encoding="utf-8", newline="") as outfile:

    reader = csv.reader(infile)
    writer = csv.writer(outfile, quoting=csv.QUOTE_MINIMAL)

    header = next(reader)
    writer.writerow(header)

    expected_cols = len(header)

    for row in reader:
        if len(row) > expected_cols:
            # Merge everything between query (index 1) and label (last 5 columns from end)
            id_val = row[0]
            # The 'label' column is always followed by 4 more columns
            label_index_from_end = 5
            query_parts = row[1:len(row) - label_index_from_end]
            query_val = ", ".join(query_parts)  # merge back into one string
            rest = row[len(row) - label_index_from_end:]
            fixed_row = [id_val, query_val] + rest
            writer.writerow(fixed_row)

        elif len(row) < expected_cols:
            fixed_row = row + [""] * (expected_cols - len(row))
            writer.writerow(fixed_row)

        else:
            writer.writerow(row)

print(f"Cleaned CSV saved to {output_file}")
```
